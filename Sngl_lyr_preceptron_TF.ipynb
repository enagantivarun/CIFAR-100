```python
#Importing tensorflow
import tensorflow as tf
```

```python
#Importing different packages like NumPy, Pandas, matplotlib

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import os
os.chdir('/Users/Varun Enaganti/Downloads')
import matplotlib.pyplot as plt 
from sklearn.utils import shuffle
```

```python
#Reading the sonar dataset
def read_dataset():
    df=pd.read_csv("sonar.all-data.csv")
    print(len(df.columns))
    X = df[df.columns[0:60]].values
    y = df[df.columns[60]]
    encoder = LabelEncoder()
    encoder.fit(y)
    y = encoder.transform(y)
    Y = one_hot_encode(y)
    return(X,Y,y)
    
```

```python
#Function to one hot encode y
def one_hot_encode(labels):
    n_labels = len(labels)
    n_unique_labels = len(np.unique(labels))
    one_hot_encode = np.zeros((n_labels,n_unique_labels))
    one_hot_encode[np.arange(n_labels),labels] = 1
    return one_hot_encode
    
```

```python
X, Y, y = read_dataset()
```

```python
#Splitting the dataset into train and test sets
X,Y = shuffle(X,Y, random_state =1)
train_x,test_x,train_y,test_y=train_test_split(X,Y, test_size=0.20, random_state=1)
print(train_x.shape)
print(test_x.shape)
print(train_y.shape)
print(test_y.shape)
```

```python
#Defining the parameters like learning rate, epoches 
learning_rate=0.1
training_epoche=1000
cost_history=np.empty(shape=[1],dtype=float)
```

```python
n_dim=X.shape[1]
n_classes=2
```

```python
#input values, weights and bias
x = tf.placeholder(tf.float32, [None, n_dim])
W = tf.Variable(tf.zeros([n_dim,n_classes]) )
b = tf.Variable(tf.zeros([n_classes]))

init = tf.global_variables_initializer()
```

```python
#Applying the softmax activation function, calculating the cost_function and optimizing using gradient descent optimizer
y_ = tf.placeholder(tf.float32, [None,n_classes])
y = tf.nn.softmax(tf.matmul(x,W)+b)
cost_function=tf.reduce_mean(-tf.reduce_sum((y_*tf.log(y)), reduction_indices=[1]))
training_step=tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost_function)
```

```python
#Intiating the session
sess=tf.Session()
sess.run(init)
mse_history =[]
```

```python
for epoche in range (training_epoche):
    sess.run(training_step, feed_dict={x:train_x,y_:train_y})
    cost = sess.run(cost_function, feed_dict={x:train_x,y_:train_y})
    cost_history=np.append(cost_history,cost)
    pred_y=sess.run(y, feed_dict={x:test_x})
    print('epoche:',epoche ,'cost:', cost)
    mse=tf.reduce_mean(tf.square(pred_y-test_y))
    mse_history.append(sess.run(mse))
    
    
print("mse:", mse_history)
plt.plot(mse_history,'ro-')
plt.show()
    
    
    

    

    
```
